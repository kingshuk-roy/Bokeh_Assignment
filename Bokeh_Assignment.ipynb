```python
# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

'''
In Bokeh, glyphs are the visual shapes or markers that represent data points on a plot.
 They can be used to create various types of plots, such as scatter plots, line plots, bar plots, and more.
  Bokeh provides a wide range of glyphs that you can use to customize your plots.

To add glyphs to a Bokeh plot, you typically use the glyph methods provided by the figure object.
Here's an example of how to add different glyphs to a scatter plot:'''

from bokeh.plotting import figure, show

# Create a figure object
p = figure()

# Add scatter plot glyphs
p.circle([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], size=10, color='red', alpha=0.5)
p.square([1, 2, 3, 4, 5], [2, 4, 6, 8, 10], size=12, color='blue', alpha=0.8)
p.triangle([1, 2, 3, 4, 5], [5, 3, 8, 6, 1], size=8, color='green', alpha=0.6)

# Display the plot
show(p)
```

```python
# Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

from bokeh.plotting import figure, show

# Create a figure object
p = figure(title='Test_Customize Plot', x_axis_label='X-axis', y_axis_label='Y-axis')

# Add data to the figure
p.circle([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], legend_label='Circle Glyph',size=10)
p.square([1, 2, 3, 4, 5], [2, 4, 6, 8, 10], legend_label='Square Glyph',size=10)

# Customize the appearance
p.title.text_color = 'blue'
p.title.text_font_size = '18pt'
p.xaxis.axis_label_text_color = 'green'
p.yaxis.axis_label_text_color = 'red'
p.legend.location = 'top_left'

# Display the plot
show(p)
```

```python
# Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?

from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
from bokeh.io import curdoc

# Create a figure object
p = figure()

# Create a data source
source = ColumnDataSource(data=dict(x=[1, 2, 3, 4, 5], y=[6, 7, 2, 4, 5]))

# Add scatter plot glyphs using the data source
p.scatter('x', 'y', source=source,size=10)

def update_data():
    # Perform some calculations or update the data in the data source
    new_data = dict(x=[1, 2, 3, 4, 5], y=[8, 3, 5, 2, 9])
    source.data = new_data

# Add the update_data() function to a periodic callback
curdoc().add_periodic_callback(update_data, 1000)  # Updates every 1 second

# Run the Bokeh server
show(p)

# bokeh serve app.py
```
